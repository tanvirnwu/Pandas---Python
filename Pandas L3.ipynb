```python
country_res = df['Country'].value_counts()
```

```python
country_res
```

```python
Countr_uses_python = Country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
```

```python
Countr_uses_python
```

```python
python_Df = pd.concat([country_res,Countr_uses_python], axis='columns',sort=False)
```

```python
python_Df
```

```python
python_Df.rename(columns={'Country':'NumofRes','LanguageWorkedWith':'Numknowspython'}, inplace=True)
```

```python
python_Df
```

```python
python_Df['Pctknowspython'] = (python_Df['Numknowspython'] / python_Df['NumofRes']) * 100
python_Df #shows the percentage of people know Python from each country
```

```python
python_Df.sort_values(by='Pctknowspython', ascending=False, inplace=True)
```
